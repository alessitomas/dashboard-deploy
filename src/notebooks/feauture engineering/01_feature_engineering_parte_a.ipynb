## PESSOA_PIPEDRIVE_Canal de Preferência

Aplica one hot encoding na coluna

```python
lost_reason_dummies = pd.get_dummies(data['PESSOA_PIPEDRIVE_Canal de Preferência'], prefix='canal_de_preferencia')
data = pd.concat([data, lost_reason_dummies], axis=1)

data
```

Coluna composta apenas por valores nulos, será dropada
